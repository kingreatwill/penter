```python
%matplotlib inline
from mpl_toolkits import mplot3d # 三维画图需要的
import matplotlib.pyplot as plt
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'TRUE' # plt绘图报错（不显示图的问题）
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

from scipy import stats
import numpy as np
```

[scipy.stats里面几乎拥有常用的分布](https://docs.scipy.org/doc/scipy/reference/stats.html)

```python
# continuous连续变量分布rv_continuous
# discrete离散变量分布rv_discrete
def drow_probability_distributions(name,rv,cdf_math):
    x = np.linspace(rv.ppf(0.0001),rv.ppf(0.9999), 100)
    plt.figure(figsize=(15,5))   #设置图片大小
    plt.subplot(1,3,1)  # pdf
    plt.plot(x, rv.pdf(x),label=name+' pdf')

    plt.legend()
    plt.subplot(1,3,2) # cdf
    plt.plot(x, rv.cdf(x),label=name+' cdf')
    plt.legend()
    plt.title(r'The probability density function for '+name+' is: '+cdf_math)
    plt.subplot(1,3,3) # ppf
    plt.plot(x, rv.ppf(x),label=name+' ppf')
    plt.legend()
```

```python
# standardized形式，https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html#scipy.stats.norm
cdf_math = r'$f(x) = \frac{exp(-x^2/2)}{\sqrt{2\pi}}$'
drow_probability_distributions("norm",stats.norm(),cdf_math)
```

```python
# standardized形式，https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gamma.html#scipy.stats.gamma
cdf_math = r'$f(x,a) = \frac{x^{a-1}e^{-x}}{\Gamma(a)} ,(x>=0,a>0)$'
drow_probability_distributions("gamma",stats.gamma(a=1.99),cdf_math)
```

```python
# standardized形式，https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.alpha.html#scipy.stats.alpha
cdf_math = r'$f(x,a) = \frac{1}{x^2\Phi(a)\sqrt{2\pi}}*exp(-\frac{1}{2}(a-1/x)^2) ,(\Phi = norm,x>0,a>0)$'
drow_probability_distributions("alpha",stats.alpha(a=3.57),cdf_math)
```
